# Topic No.1:
    Scraping and Crawling the data from IMDB BOLLYWOOD

```python
import pandas as pd
from bs4 import BeautifulSoup
import requests

url = 'https://www.imdb.com/list/ls025929404/?sort=list_order,asc&mode=detail&page={}'
df=pd.DataFrame()
for url_no in range(1,3,1):
    r = requests.get(url.format(url_no))
    print(url.format(url_no))
    c = r.content
    soup= BeautifulSoup(c,'html.parser')
    all=soup.select(".lister-item.mode-detail")
    names=[item.select_one('.lister-item-image img[alt]')['alt'] for item in all]
    movie=[item.select_one('.text-muted.text-small>a').text.strip() for item in all]
    Profession=[item.select_one('.text-muted.text-small').next_element.strip() for item in all]
    Image=[item.select_one('.lister-item-image img[src]')['src'] for item in all]
    Paragraph=[item.select_one('.text-muted.text-small+p').text.strip() for item in all]


    df1=pd.DataFrame({"Name":names,"Star_Introduction":Paragraph,"Image":Image,"Profession":Profession,"Movie":movie})
    df=df.append(df1,ignore_index=True)
```

```python
import numpy as np
import pandas as pd
from bs4 import BeautifulSoup
import requests
r = requests.get('https://www.imdb.com/search/keyword/?page=1&keywords=behind-the-scenes&explore=keywords&mode=detail&ref_=kw_nxt&sort=moviemeter,asc')
c = r.content
soup = BeautifulSoup(c,'html.parser')
last= soup.find('div',{'class':'desc'})
last = last.text.replace('to','').replace('of','').replace('titles','')
last = last.replace(' \n            ','\n').replace('\n             ','\n')
last = last.replace(' \n    | Next »\n','').replace(',','')
last = last.split('\n')

last= last[1:4]
results = [int(i) for i in last]
last = results[2]/results[1]
last = round(last)
print(last)


url = 'https://www.imdb.com/search/keyword/?page={}&keywords=behind-the-scenes&explore=keywords&mode=detail&ref_=kw_nxt&sort=moviemeter,asc'

df=pd.DataFrame()
#for url_no in range(1, last+1):
for url_no in range(1,last):
    r = requests.get(url.format(url_no))
    print(url.format(url_no))
    c = r.content
    soup= BeautifulSoup(c,'html.parser')
    all=soup.select(".lister-item.mode-detail ")
    
    names=[item.select_one('.lister-item-image img[alt]')['alt'] for item in all]
    
    try:
        year = [item.select_one('.lister-item-year').text.strip() for item in all]
    except:
        year = None
    
    try:
        star=[item.select_one('.text-muted.text-small>a').text.strip() for item in all]
    except:
        star = None
        
    try:
        genre_minute = [item.select_one('.text-muted.text-small').text.strip() for item in all]
    except:
        genre_minute = None
    try: 
        rating = [item.select_one('.ratings-imdb-rating').text.strip() for item in all]
    except:
        rating = None
    try: 
        all_data = [item.select_one('.lister-item-content').text.strip() for item in all]
    except:
        all_data = None
    try:
        Vote=[item.select_one('.text-muted.text-small+p').text.strip() for item in all]
    except:
        Vote = None
    try:
        Image=[item.select_one('.lister-item-image img[loadlate]')['loadlate'] for item in all]
    except:
        Image= None
        
    
    df1=pd.DataFrame({"Movie_Name":names,"Star":star,"Year":year,'Ratings':rating,'Genre_Minute':genre_minute,'all_data':all_data, 'Vote':Vote, 'Image':Image})
    df=df.append(df1,ignore_index=True)

df['Genre'] = df['Genre_Minute'].apply(lambda x:x.split('\n')[-1])
df['Minutes'] = df['all_data'].apply(lambda x : x .split(' min\n|\n\n')[0].split('\n')[-1])

print(len(df))
```
